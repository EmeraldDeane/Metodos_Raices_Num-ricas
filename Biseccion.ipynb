```python
# Graficar la función y aplicar el método de bisección
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 7*x + 6

# Graficar la función para justificar el intervalo
x_vals = np.linspace(-5, 5, 400)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals, label='f(x)')
plt.axhline(0, color='gray', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de f(x) = x^3 - 7x + 6')
plt.grid(True)
plt.legend()
plt.show()

# Método de bisección
def biseccion(f, a, b, tol=1e-5):
    if f(a) * f(b) >= 0:
        raise ValueError("La función no cambia de signo en el intervalo dado.")
    iteraciones = 0
    while (b - a)/2 > tol:
        c = (a + b) / 2
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iteraciones += 1
    return (a + b) / 2, iteraciones

# Elegimos el intervalo [0, 2] por el cambio de signo
a, b = 0, 2
raiz, n_iter = biseccion(f, a, b)
print(f"Raíz aproximada: {raiz:.5f}")
print(f"Iteraciones requeridas: {n_iter}")
```
